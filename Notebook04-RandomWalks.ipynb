### One-dimensional random walks
Suppose that you decided to repeatedly gamble on the result of a fair coin flip, with the bet being \$1 on whether or not the result is heads. Should you play this game 100 times in a row, your balance would almost certainly teeter upwards and downwards throughout the process; this fluctuation precisely corresponds to one of the most studied phenomena in probability: the **random walk**.

```python
import random
import numpy as np

# Return the entire trajectory from randomly walking either left or right
# a given number of steps
def randomWalk(numSteps):
    positions=np.zeros(numSteps)
    pos = 0
    positions[0] = pos
    for j in xrange(1,numSteps):
        flip = random.randint(0,1)
        if flip == 0: pos -= 1
        else:         pos += 1
        positions[j] = pos
    return positions

# Return the final position from randomly walking either left or right a 
# given number of steps
def randomWalkResult(numSteps):
    pos = 0
    for j in xrange(1,numSteps):
        flip = random.randint(0,1)
        if flip == 0: pos -= 1
        else:         pos += 1
    return pos
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

numSteps = 100
positions = randomWalk(numSteps)
plt.plot(positions,'ko-')
```

```python
# Return the entire trajectory from randomly walking in one of four
# directions for a given number of steps
def randomWalk2D(numSteps):
    xPositions=np.zeros(numSteps)
    yPositions=np.zeros(numSteps)
    xPos=yPos=0
    xPositions[0] = xPos
    yPositions[0] = yPos
    for j in xrange(1,numSteps):
        flip = random.randint(0,3)
        if   flip == 0: xPos += 1 # walk right
        elif flip == 1: yPos += 1 # walk up
        elif flip == 2: xPos -= 1 # walk left
        else:           yPos -= 1 # walk down
        xPositions[j] = xPos
        yPositions[j] = yPos
    return xPositions, yPositions

# Return the final result from randomly walking in one of four 
# directions for a given number of steps
def randomWalk2DResult(numSteps):
    xpos=ypos=0
    for j in xrange(1,numSteps):
        flip = random.randint(0,3)
        if   flip == 0: xPos += 1 # walk right
        elif flip == 1: yPos += 1 # walk up
        elif flip == 2: xPos -= 1 # walk left
        else:           yPos -= 1 # walk down
    return xPos, yPos
```

```python
numSteps=10000
xPositions, yPositions = randomWalk2D(numSteps)

cm = plt.get_cmap('winter')

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_color_cycle([cm(1.*i/(numSteps-1)) for i in xrange(numSteps-1)])
for j in xrange(numSteps-1):
    plt.plot(xPositions[j:j+2],yPositions[j:j+2])
```

```python
# Return the entire trajectory from randomly walking in one of six
# directions for a given number of steps
def randomWalk3D(numSteps):
    xPositions=np.zeros(numSteps)
    yPositions=np.zeros(numSteps)
    zPositions=np.zeros(numSteps)
    xPos=yPos=zPos=0
    xPositions[0] = xPos
    yPositions[0] = yPos
    zPositions[0] = zPos
    for j in xrange(1,numSteps):
        flip = random.randint(0,5)
        if   flip == 0: xPos += 1 # walk right
        elif flip == 1: yPos += 1 # walk up
        elif flip == 2: xPos -= 1 # walk left
        elif flip == 3: yPos -= 1 # walk down
        elif flip == 4: zPos += 1 # walk into the plane
        else:           zPos -= 1 # walk out of plane
        xPositions[j] = xPos
        yPositions[j] = yPos
        zPositions[j] = zPos
    return xPositions, yPositions, zPositions

# Return the final position after randomly walking in one of six 
# directions for a given number of steps
def randomWalk3DResult(numSteps):
    xPos=yPos=zPos=0
    for j in xrange(1,numSteps):
        flip = random.randint(0,5)
        if   flip == 0: xPos += 1 # walk right
        elif flip == 1: yPos += 1 # walk up
        elif flip == 2: xPos -= 1 # walk left
        elif flip == 3: yPos -= 1 # walk down
        elif flip == 4: zPos += 1 # walk into the plane
        else:           zPos -= 1 # walk out of plane
    return xPos, yPos, zPos
```

```python
numSteps=10000
xPositions, yPositions, zPositions = randomWalk3D(numSteps)

import mpl_toolkits.mplot3d.axes3d as p3
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_color_cycle([cm(1.*i/(numSteps-1)) for i in range(numSteps-1)])
#scatter = ax.scatter(xPositions,yPositions,zPositions,c=range(numSteps))
for j in xrange(numSteps-1):
    ax.plot(xPositions[j:j+2],yPositions[j:j+2],zPositions[j:j+2])
```
