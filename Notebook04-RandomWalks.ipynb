### Spread
Our previous analysis has focused on quantifying several different notions of what the 'typical' value of a random process is from the perspective of minimizing different norms of the weighted difference of this characteristic value with the possible outcomes. For example, our first set of discussions connected the **mean** and the **two-norm** and the **median** and the **one-norm**, and later discussions involved the introduction of the notion of an **expectation** of a random process, which we interpreted as computing the **mean** behaviour (with the simplest example being that, if we flip a fair coin and win a dollar if it turns up heads, we can expect to win fifty cents on average). 

However, when we simulated some of the random processes discussed in class (both physically and in much more detail via simple Python code), we saw that the averages from our experimental results could, for the coin-flipping game with large, rare payoffs, vary significantly from the mathematical expectation, even after performing thousands of trials. We then did several straight-forward calculations of the **standard deviation** and saw that there seemed to be a connection between large standard deviations and mismatches in our experimental and theoretical average behavior. 

The purpose of this note is to discuss several different interpretations of how much a random event tends to deviate from its 'typical' result, often referred to as the **spread**, with the **standard deviation** being the textbook example. In order to do so, we will begin by describing an incredibly important random process whose mean behaviour is always zero, but whose spread steadily increases.

### One-dimensional random walks 
Suppose that you decided to repeatedly gamble on the result of a fair coin flip, with the bet being that you gain \$1 if heads comes up, and lose \$1 if tails comes up. Should you play this game 100 times in a row, your balance would almost certainly teeter upwards and downwards in an erratic manner. In fact, this fluctuation has been intensely studied by both mathematicians and physicists for over a century and is typically referred to as a **random walk**, with one interpretation being a random process where someone randomly chooses to take either one step forward or one step backwards a large number of times. 

With this interpretation, the first question might be, after taking $n$ random steps, where would you end up on average? Perhaps unsurprisingly, since the expected change from each step is zero, the random walk will, on average, end up where it started. But, clearly it's not the case that we *always* end the random walk at the origin, and so a great follow-up question might be:

> How far away do the $n$ random steps tend to take us from where we started?

This question is an example of inquiring about the **spread** of a random variable about its average, and the answer in the case turns out to be simple if we only measure the **average distance squared** from the mean (known as the **variance**, which is the **square** of the **standard deviation**), but relatively complicated if we only measure the **average distance** from the mean (known as the **Mean Absolute Deviation (MAD)**).

### Computing the variance
We can define the **variance** of a random variable $X$ as the **average squared distance from the average**,

$$ \text{Var}[X] = E[(X-E[X])^2], $$

where $E[X]$ is the **expected value** (the average behaviour) of the random variable $X$, $(X - E[X])^2$ is the square of the distance from the average, and the variance is the average value of this squared distance. Furthermore, the **standard deviation** is defined as the square-root of the variance, and so we might write

$$ \text{SD}[X] = \sqrt{\text{Var}[X]} = \sqrt{E[(X-E[X])^2]}. $$

As an example, the variances after either one or two steps of the random walk can be quickly computed after constructing the following tables of possible results:

We can see that, after only one step, there are two possible results (-1 or +1), while after two steps, there are three possibilities (-2,0, and +2). The variances of the random walk after one and two steps can then computed as

$$ \begin{eqnarray}
   \text{Var}[W_1] &=& \frac{1}{2}(1-0)^2 + \frac{1}{2}(-1-0)^2 = 1, \\
   \text{Var}[W_2] &=& \frac{1}{4}(2-0)^2 + \frac{1}{2}(0-0)^2 + \frac{1}{4}(2-0)^2 = 2. 
   \end{eqnarray} $$

While the nested application of expectations might be initially difficult to interpret, once we have computed the expectation of our random variable $X$, say $\mu = E[X]$, we can simply substitute this mean into the above formula to find that

$$ \text{Var}[X] = E[(X-\mu)^2], $$

and, for the case of $n$ steps of our random walk, say, $W_n$, we have that $E[W_n] = 0$, as so the variance, in this case, simplifies to

$$ \text{Var}[W_n] = E[W_n^2]. $$

In fact, as was argued in class, the variance of the final location after randomly taking $n$ steps is $n$, and we boiled the fundamental reason down to the expected change in the squared distance from zero after taking one random step from an arbitrary position $k$:

  $$\left(\frac{1}{2}((k+1)-0)^2 + \frac{1}{2}((k-1)-0)^2\right) - (k-0)^2 = \frac{(k+1)^2+(k-1)^2}{2} - k^2 = 1. $$
  
Thus, it can be expected that each step of the random walk increases the variance by $1$, and it holds that, after $n$ steps,

$$ \text{Var}[W_n] = E[W_n^2] = n. $$

This would suggest that a large number of people who had randomly taken $n$ steps would tend to have an average squared distance of $n$ from where they started (even though the average of all of their positions would tend to be zero).

### Computing the Mean Absolute Deviation with binomial coefficients
Similarly, the **Mean Absolute Deviation** of a random variable about some central value $\gamma$ (with $\gamma$ typically being set to $E[X]$) is defined as the **average distance** from $\gamma$, or

$$ \text{MAD}[X] = E[|X-\gamma|]. $$

While this formula is simpler than the definition of the variance, it turns out to often be more difficult to work with (for reasons most easily explained with basic calculus); we shall now see that it is significantly more difficult to compute the Mean Absolute Deviation of the random walk than it was to compute its variance.

While we can reuse the tables from above and see that the Mean Absolute Deviation about zero after one and two steps of our random walk would respectively be

$$ \begin{eqnarray}
   \text{MAD}[W_1] &=& \frac{1}{2}|1-0| + \frac{1}{2}|-1-0| = 1, \\
   \text{MAD}[W_2] &=& \frac{1}{4}|2-0| + \frac{1}{2}|0-0| + \frac{1}{4}|2-0| = 1, 
   \end{eqnarray} $$

calculations for other numbers of steps (e.g., $\text{MAD}[W_3]=1.5$) lead to results that are, at first, difficult to easily explain and predict.

Thankfully, just as was the case for the variance of the random walk, it is possible to boil down the fundamental behaviour of the MAD of the random walk by considering the expected change in the absolute distance from zero after taking one step from a fixed location. For example, supposing that we have taken some arbitrary number of random steps from zero and ended up at position $5$, then the expected change in absolute distance after one more step would be

$$ \left(\frac{1}{2}|(5+1)-0| + \frac{1}{2}|(5-1)-0|\right) - |5-0| = 0. $$

But if we instead considered the expected change in absolute distance if our random walk was currently at the origin, we would see

$$ \left(\frac{1}{2}|(0+1)-0| + \frac{1}{2}|(0-1)-0|\right) - |0-0| = 1. $$

And, in fact, we would see that, unless the walk happens to have been at the origin, taking one more step does not change the expected distance from zero, but, if the walk is currently at the origin, the next step increases the distance from zero by 1. Thus, we can express the increase in Mean Absolute Deviation from the $n$'th step to the $n+1$'th step as

$$ \text{MAD}[W_{n+1}] = \text{MAD}[W_n] + \left(P[W_n=0]*1 + P[W_n \neq 0]*0\right) = \text{MAD}[W_n] + P[W_n=0], $$

where we have used the notation $P[W_n=0]$ to mean "the probability that $W_n$ is zero" and $P[W_n \neq 0]$ to mean "the probability that $W_n$ is **not** zero".

Since $\text{MAD}[W_0]=0$ (as the walk starting at zero is guaranteed to be at zero if no steps have been taken), then we can write the MAD after $n$ random steps as the sum of the likelihoods that each of the previous steps had landed back at the origin:

$$ \text{MAD}[W_n] = \sum_{k=1}^{n-1} P[W_k = 0]. $$

Furthermore, since it is impossible for an odd number of steps to end at zero (since an equal number of forward and backward steps implies an even number of total steps), the above summation can in fact be simplified to only contain the even terms, as $P[W_k=0]$ will be zero whenever $k$ is odd.

#### Binomial coefficients
But to determine the probability that the random walk lands at the origin after an even number of steps, it is useful to make use of one of the standard tools of combinatorial analysis: the **choose function**, whose values are often called **binomial coefficients**

$$ \binom{n}{k} = \frac{n*(n-1)*\cdots*(n-(k-1))}{k!}, $$

which can be interpreted as the number of different ways that $k$ objects can be uniquely selected from $n$ objects if the order that they are selected is ignored. In particular, the numerator of the definition consists of $k$ terms and is equivalent to the number of unique ways that the $k$ objects can be selected if the ordering is taken into account, and the denominator is precisely the number of different ways that $k$ objects can be ordered.

As an example, if we consider four letters, say, $a, b, c, d$, then the different ways to select two letters out of these four are

$$ \begin{array}{cccc} ab & ba & ca & da \\ 
                       ac & bc & cb & db \\
                       ad & bd & cd & dc \end{array}, $$
                       
and it can be seen that, of these twelve items, each pair (e.g., $(a,b)$), appears twice (i.e., as $ab$ and $ba$), and thus there are 6 unique ways to select two items from four if the ordering is ignored. This perfectly aligns with our above formula for $\binom{4}{2}$, as 

$$ \binom{4}{2} = \frac{4*3}{2*1} = \frac{12}{2} = 6. $$

We can now answer the question of how likely it is that an even number, say $2k$, of steps of our random walk would end at the origin, as it is equivalent to asking the number of different ways that $k$ of the $2k$ steps could have been backwards steps, which is equivalent to 

$$\binom{2k}{k} = \frac{2k*(2k-1)*\cdots*(k+1)}{k*(k-1)*\cdots*1},$$

divided by the total number of (equally-likely) possibilities, which is $2^{2k}$, since there are $2k$ independent choices of whether to step forward or backward.

We can thus write down the two ways in which the MAD can change (or not change) from one step to the next:

$$ \begin{eqnarray}
   \text{MAD}[W_{2k}] &=& \text{MAD}[W_{2k-1}] + P[W_{2k-1}=0] = \text{MAD}[W_{2k-1}],\;\text{and} \\
   \text{MAD}[W_{2k+1}] &=& \text{MAD}[W_{2k}] + P[W_{2k}=0] = \text{MAD}[W_{2k}] + \frac{1}{2^{2k}}\binom{2k}{k}. \end{eqnarray} $$

### Interquartile ranges, outliars, and box plots
While the variance and Mean Absolute Deviation are useful ways of describing the spread of a dataset about its mean, it is also popular to use "one-norm" style analysis, which we argued makes use of medians, in order to classify points according to the **quartiles**. In particular, the median of a dataset is a value such that at least half of the samples are greater than or equal to the value, and at least half are less than or equal to the value. We generalize this notion by dividing the dataset into four pieces using three **quartiles** rather than dividing it into two pices with the **median**. In particular, the **lower quartile** can be defined as the median of the data which is less than or equal to the median, the **middle quartile** is simply the median, and the **upper quartile** is the median of the data greater than or equal to the median. For example, for the (sorted) dataset 

$$ (-120, -50, -40, 0, 10, 20, 30, 40, 50, 100, 150) $$

the median / middle-quartile is $20$, the lower quartile is the median of $(-120,-50,-40,0,10,20)$, which is $-20$, and the upper quartile is the median of $(20,30,40,50,100,150)$, which is $45$.

The distance between the lower and upper quartiles is called the **Interquartile Range**, and, for the previous dataset, this would be equal to $45-(-20)=65$. It is then common to call a datapoint an **outlier** if it is either:

1. Greater than the upper quartile plus 1.5 times the IQR, or
2. Less than the lower quartile minus 1.5 times the IQR.

For the above dataset, these requirements would classify a point as an oulier if it is either less than $-20-1.5*65=-117.5$ or greater than $45+1.5*65=142.5$. As can be seen, $-120$ satisfies the first condition and $150$ satisfies the second, and so each would be called an outlier. This analysis is frequently visually displayed via what is referred to as a **Tukey box plot**, which displays the three quartiles via a box, the two ranges of data points lying outside of the box but within 1.5 IQRs of the lower or upper quartiles via two "*whiskers*", and explicitly draws each outliar with a single point. An example of a box plot for the discussed dataset is given below.

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [-120,-50,-40,0,10,20,30,40,50,100,150]

fig = plt.figure()
plt.boxplot(x, 
            notch=False, # box instead of notch shape 
            sym='rs',    # red squares for outliers
            vert=True)   # vertical box aligmnent
t = plt.title('Box plot')
plt.show()
```

### Random walks in two and three dimensions
As an aside, it is of course also possible to define and simulate random walks which take place in more than one spatial dimension. In the case of two dimensions, the usual convention is to generalize from the 1D rule of either moving upward (+1) or downward (-1) with equal likelihood to allowing for the movement to either move up one, down one, right one, or left one with a probability of $1/4$ each. Two simple routines for simulating this random process are defined in Python below, with the first routine returning the entire trajectory of the walk, and the latter only computing the final location of the walk (which runs significantly faster):

```python
# Return the entire trajectory from randomly walking in one of four
# directions for a given number of steps
def randomWalk2D(numSteps):
    xPositions=np.zeros(numSteps)
    yPositions=np.zeros(numSteps)
    xPos=yPos=0
    xPositions[0] = xPos
    yPositions[0] = yPos
    for j in xrange(1,numSteps):
        flip = random.randint(0,3)
        if   flip == 0: xPos += 1 # walk right
        elif flip == 1: yPos += 1 # walk up
        elif flip == 2: xPos -= 1 # walk left
        else:           yPos -= 1 # walk down
        xPositions[j] = xPos
        yPositions[j] = yPos
    return xPositions, yPositions

# Return the final result from randomly walking in one of four 
# directions for a given number of steps
def randomWalk2DResult(numSteps):
    xpos=ypos=0
    for j in xrange(1,numSteps):
        flip = random.randint(0,3)
        if   flip == 0: xPos += 1 # walk right
        elif flip == 1: yPos += 1 # walk up
        elif flip == 2: xPos -= 1 # walk left
        else:           yPos -= 1 # walk down
    return xPos, yPos
```

We can now run the first procedure for 10,000 steps and visualize the result by plotting the positions over time with a color scheme which starts at dark blue and gradually transitions to light green at the last step.

```python
cm = plt.get_cmap('winter')

numSteps=10000
xPositions, yPositions = randomWalk2D(numSteps)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_color_cycle([cm(1.*i/(numSteps-1)) for i in xrange(numSteps-1)])
for j in xrange(numSteps-1):
    plt.plot(xPositions[j:j+2],yPositions[j:j+2])
```

But if we rerun the simulation, we will, of course, get a very different pattern for the walk (and, for those of you making use of the interactive notebook, you may want to rerun the simulation a few times to get a feel for the behavior).

Lastly, nothing would stop us from defining a three-dimensional random walk which would randomly choose between one of six directions at each step (either forward or backward in one of the three dimensions). The following few cells of the notebook define procedures for simulating a 3D random walk and then perform two simulations and plot the color-coded results.

```python
# Return the entire trajectory from randomly walking in one of six
# directions for a given number of steps
def randomWalk3D(numSteps):
    xPositions=np.zeros(numSteps)
    yPositions=np.zeros(numSteps)
    zPositions=np.zeros(numSteps)
    xPos=yPos=zPos=0
    xPositions[0] = xPos
    yPositions[0] = yPos
    zPositions[0] = zPos
    for j in xrange(1,numSteps):
        flip = random.randint(0,5)
        if   flip == 0: xPos += 1 # walk right
        elif flip == 1: yPos += 1 # walk up
        elif flip == 2: xPos -= 1 # walk left
        elif flip == 3: yPos -= 1 # walk down
        elif flip == 4: zPos += 1 # walk into the plane
        else:           zPos -= 1 # walk out of plane
        xPositions[j] = xPos
        yPositions[j] = yPos
        zPositions[j] = zPos
    return xPositions, yPositions, zPositions

# Return the final position after randomly walking in one of six 
# directions for a given number of steps
def randomWalk3DResult(numSteps):
    xPos=yPos=zPos=0
    for j in xrange(1,numSteps):
        flip = random.randint(0,5)
        if   flip == 0: xPos += 1 # walk right
        elif flip == 1: yPos += 1 # walk up
        elif flip == 2: xPos -= 1 # walk left
        elif flip == 3: yPos -= 1 # walk down
        elif flip == 4: zPos += 1 # walk into the plane
        else:           zPos -= 1 # walk out of plane
    return xPos, yPos, zPos
```
